```python
import yfinance as yf #Yahoo's Finance Library
import pandas as pd
import numpy as np

def generate_persona_scores():
  tickers = [
        'CRSP', 'EDIT', 'NTLA', 'PACB', 'COIN', 'MQ', 'HOOD', 'NU',
        'TSLA', 'RIVN', 'NIO', 'LCID', 'NVDA', 'AMD', 'PLTR', 'PATH',
        'AI', 'SNOW', 'PYPL', 'SHOP', 'ROKU', 'TDOC', 'ZM', 'RBLX', 'U'
  ]

  #Step 1: Computing Features

  data = []
  for ticker in tickers:
    try:
        t = yf.Ticker(ticker)
        info = t.info
        hist = t.history(period='1y')

        # Feature 1: Revenue Growth
        rev_growth = info.get('revenueGrowth', np.nan)

        # Feature 2: Debt-to-Equity
        debt_equity = info.get('debtToEquity', np.nan)

        # Feature 3: Moving Average Trend
        if len(hist) >= 200:
            ma_50 = hist['Close'].rolling(50).mean().iloc[-1]
            ma_200 = hist['Close'].rolling(200).mean().iloc[-1]

            if pd.isna(ma_50) or pd.isna(ma_200):
                ma_trend = np.nan  # Use NaN if MAs can't be calculated
            else:
                ma_trend = 1 if ma_50 > ma_200 else 0
        else:
            ma_trend = np.nan  # Use NaN if not enough history

        data.append([ticker, rev_growth, debt_equity, ma_trend])
    except Exception as e:
        print(f" Could not process {ticker}: {e}")
        data.append([ticker, np.nan, np.nan, np.nan])


  df = pd.DataFrame(data, columns=['ticker', 'rev_growth', 'debt_equity', 'ma_trend'])
  #print(df)

  #Step 2: Normalizing Features and CLeaning Data

  df['rev_growth'] = df['rev_growth'].fillna(df['rev_growth'].median())
  df['debt_equity'] = df['debt_equity'].fillna(df['debt_equity'].median())
  df['ma_trend'] = df['ma_trend'].fillna(0)


  def normalize(series):
      min_val = series.min()
      max_val = series.max()
      if min_val == max_val:
          return pd.Series([0.5] * len(series), index=series.index)
      return (series - min_val) / (max_val - min_val)


  df['rg_score'] = normalize(df['rev_growth'])
  df['de_score'] = 1 - normalize(df['debt_equity'])
  df['ma_score'] = df['ma_trend']

  df['score'] = (df['rg_score'] + df['de_score'] + df['ma_score']) / 3

  #print(df[['ticker', 'rg_score', 'de_score', 'ma_score', 'score']])

  #Step 3: Creating CSV Ouput

  output = df[['ticker', 'score']].copy()
  output['persona'] = "Cathie Wood"
  output.to_csv('persona_output.csv', index=False)
  print(output)

generate_persona_scores()
```
