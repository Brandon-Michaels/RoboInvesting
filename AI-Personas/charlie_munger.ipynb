Charlie Munger is a man who prefers to invest in high-quality businesses that have simple economics (nothing too complex). He also wants to ensure that he does not take big losses at all and would rather prefer small gains. He prioritizes long-term success over short-term pain.

For stocks, Munger likes to have a concentrated position on compaines over a broad diversification across many companies. He will occasionally trade with unknown companies (foreign and emerging mostly) when they offer economics he likes. He also ensures that the business have reliable ownership, not ones that make him question. He wanted to make decisions with as little bias as possible.

He wanted the signals to have "moat" through pricing power, brand strength, and network effects that would help yield high and sustainable returns on invested capital.

```python
# Import Necessary Libraries

import yfinance as yf # Yahoo's Finance Library
```

```python
def get_ticker_data(tickers):
  tickerList = [ticker for ticker, all in tickers.tickers.items()]
  for ticker, all in tickers.tickers.items():
    print(ticker)
  return yf.download(tickerList, period='4mo')
```

```python
tickers = yf.Tickers('DJCO BAC WFC USB BABA AAPL WMT KO PG CL UL V BRK-B USB UNP WM CMI ADBE ADP INTU COST CL UL')

get_ticker_data(tickers)
```

Week 2 Goal: generate a numerical score for each stock my persona tracks.

```python
import pandas as pd
# First three indicators to look at: P/B, ROE, Debt-to-Equity

ind = pd.DataFrame(columns=["Ticker", "P/B", "ROE", "DOE"])

for name, obj in tickers.tickers.items():
  # Price-To-Book
  pb = obj.info['priceToBook']
  # Return on Equity
  financials = obj.financials
  balance_sheet = obj.balance_sheet

  net_income = financials.loc["Net Income"][0]
  equity = balance_sheet.loc["Stockholders Equity"][0]

  roe = net_income / equity
  # Debt-on-Equity
  liab = balance_sheet.loc["Total Liabilities Net Minority Interest"][0]

  doe = liab / equity

  ind.loc[len(ind)] = [name, pb, roe, doe]

ind['P/B_norm'] = 1 - (ind['P/B'] - ind['P/B'].min()) / (ind['P/B'].max() - ind['P/B'].min())
ind['ROE_norm'] = (ind['ROE'] - ind['ROE'].min()) / (ind['ROE'].max() - ind['ROE'].min())
ind['DOE_norm'] = 1 - (ind['DOE'] - ind['DOE'].min()) / (ind['DOE'].max() - ind['DOE'].min())

ind
```

```python
# Generate the score

ind["score"] = 0.4 * ind['ROE_norm'] + 0.30 * ind['P/B_norm'] + 0.30 * ind['DOE_norm']

ind
```

```python
df = ind[['Ticker', 'score']]
df['persona'] = 'Charlie Munger'

df.to_csv('persona_output.csv', index=False)
```
