```python
def calculate_pabrai_score(ticker_symbol):
    """
    Calculate Mohnish Pabrai-inspired investment score (0-1)
    Higher scores indicate better alignment with Pabrai's investment philosophy
    """
    try:
        ticker = yf.Ticker(ticker_symbol)
        info = ticker.info
        hist = ticker.history(period="1y")

        if hist.empty or not info:
            return 0.5  # Default neutral score for unavailable data

        scores = []
        weights = []

        # Indicator 1: Valuation Score (P/E and P/B) - Pabrai seeks undervalued companies
        pe_ratio = info.get('trailingPE', None)
        pb_ratio = info.get('priceToBook', None)

        if pe_ratio and pe_ratio > 0:
            # Lower P/E is better - normalize with cap at 30
            pe_score = max(0, 1 - (pe_ratio / 30))
            scores.append(pe_score)
            weights.append(0.3)

        if pb_ratio and pb_ratio > 0:
            # Lower P/B is better - normalize with cap at 5
            pb_score = max(0, 1 - (pb_ratio / 5))
            scores.append(pb_score)
            weights.append(0.2)

        # Indicator 2: Profitability Score - Pabrai likes strong profit margins
        profit_margins = info.get('profitMargins', None)
        if profit_margins:
            # Higher profit margins are better - normalize with cap at 0.3 (30%)
            profit_score = min(1, profit_margins / 0.3)
            scores.append(profit_score)
            weights.append(0.25)

        # Indicator 3: Debt Safety Score - Pabrai avoids excessive debt
        debt_to_equity = info.get('debtToEquity', None)
        if debt_to_equity:
            # Lower debt/equity is better - normalize with cap at 2
            debt_score = max(0, 1 - (debt_to_equity / 2))
            scores.append(debt_score)
            weights.append(0.15)

        # Indicator 4: Business Stability Score - Pabrai prefers stable, understandable businesses
        beta = info.get('beta', None)
        if beta:
            # Lower beta (less volatility) is better for Pabrai's low-risk approach
            # Normalize: beta of 0.5 gets max score, beta of 1.5 gets min score
            beta_score = max(0, min(1, (1.5 - beta) / 1.0))
            scores.append(beta_score)
            weights.append(0.1)

        # Calculate weighted average score
        if scores and weights:
            weighted_score = sum(s * w for s, w in zip(scores, weights)) / sum(weights)
            return round(weighted_score, 4)
        else:
            return 0.5

    except Exception as e:
        print(f"Error calculating score for {ticker_symbol}: {str(e)}")
        return 0.5

def get_pabrai_stock_signals():
    """Generate Mohnish Pabrai-inspired investment signals for tracked stocks"""

    # Pabrai's typical investment universe - quality companies with value characteristics
    tickers = [
        'JNJ',    # Johnson & Johnson - stable healthcare, strong margins
        'MRNA',   # Moderna - biotech with potential, but higher risk
        'NKE',    # Nike - strong brand, understandable business
        'AAPL',   # Apple - quality tech, strong cash flow
        'SHW',    # Sherwin-Williams - simple business model
        'BIDU',   # Baidu - Chinese value play
        'TCEHY',  # Tencent - quality Chinese tech
        'INTC',   # Intel - potential turnaround, value characteristics
        'RL',     # Ralph Lauren - brand value
        'AMZN',   # Amazon - growth at reasonable price potential
        'SYK',    # Stryker - medical devices, stable business
        'YUM',    # Yum! Brands - simple restaurant business
        'GOOGL',  # Alphabet - quality tech, strong margins
        'LVS',    # Las Vegas Sands - higher risk, potential reward
        'AMD',    # AMD - more volatile, growth potential
        'PFE',    # Pfizer - stable pharma, value characteristics
        'ABBV',   # AbbVie - pharma with strong dividends
        'COST',   # Costco - excellent business model
        'GE',     # General Electric - turnaround potential
        'CVX',    # Chevron - energy value play
        'T',      # AT&T - telecom with value characteristics
        'SPG',    # Simon Property Group - real estate value
        'RCL',    # Royal Caribbean - higher risk recovery play
        'PYPL',   # PayPal - fintech with growth potential
    ]

    signals = []

    print("Calculating Mohnish Pabrai investment scores...")
    for ticker in tickers:
        score = calculate_pabrai_score(ticker)
        signals.append({
            'ticker': ticker,
            'score': score,
            'persona': 'Mohnish Pabrai'
        })
        print(f"{ticker}: {score:.4f}")

    # Create DataFrame and sort by score (highest first)
    df = pd.DataFrame(signals)
    df = df.sort_values('score', ascending=False)

    return df

def generate_persona_output():
    """Generate the final output CSV in required format"""
    df = get_pabrai_stock_signals()

    # Save to CSV
    output_filename = 'persona_output.csv'
    df.to_csv(output_filename, index=False)

    print(f"\nTop 5 Pabrai Recommendations:")
    print(df.head())
    print(f"\nFull results saved to {output_filename}")

    return df

# Execute the analysis
if __name__ == "__main__":
    results = generate_persona_output()
```
